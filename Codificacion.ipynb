# Codificación de variables categóricas y binarias
Este cuaderno aplica codificaciones en one-hot a las variables `Gender`, `CAEC`, `CALC`, `MTRANS` y `NObeyesdad`, y transforma las variables binarias `family_history_with_overweight`, `FAVC`, `SMOKE` y `SCC` a valores 0/1.


```python
import pandas as pd

DATA_PATH = 'ObesityDataSet_raw_and_data_sinthetic.csv'
dataset = pd.read_csv(DATA_PATH)
dataset.head()
```

```python
categorical_cols = ['Gender', 'CAEC', 'CALC', 'MTRANS', 'NObeyesdad']
binary_cols = ['family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']

encoded_df = dataset.copy()
encoded_df[binary_cols] = encoded_df[binary_cols].replace({'yes': 1, 'no': 0})
encoded_df = pd.get_dummies(encoded_df, columns=categorical_cols, drop_first=False)

encoded_df.head()
```
